```python
import os

json_dir_list = []  # zip 파일이 있는 경로를 담을 리스트
with open('JSON_dir.txt', 'r') as f :
  json_dir_list = f.readlines()  # 텍스트의 내용을 리스트에 담기 - 줄바꿈이 \n으로 문자열 끝에 붙음

json_dir_list = [path.strip() for path in json_dir_list]  # 문자열의 공백문자 제거
print(json_dir_list)  # 경로 리스트 확인
```

```python
print(len(json_dir_list))
print('\n'.join(json_dir_list))
```

```python
# 훈련, 테스트용 json 데이터 경로
suv_json_dirs = [path for path in json_dir_list if 'SUV' in path]
train_suv_json = [path for path in suv_json_dirs if 'Training' in path]
val_suv_json = [path for path in suv_json_dirs if 'Validation' in path]
```
