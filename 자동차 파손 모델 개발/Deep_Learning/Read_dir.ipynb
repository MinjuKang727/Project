# 폴더 경로 수집

````python
# 로컬 드라이브용 코드
import os

f_dic = {'160. 차량파손 이미지 데이터':{}, '자동차 차종-연식-번호판 인식용 영상':{}}  # 상하위 폴더를 작성할 딕셔너리
ignore_f = ['.zip','.irx', 'irx1']  # 폴더 정보 수집시, 무시할 확장자명
end_f = ['.jpg', '.xls', 'json']  # 폴더의 제일 마지막에 있는 파일의 확장자명
root_dir = "C:/Users/AI/Downloads/"
f_list = ['160. 차량파손 이미지 데이터', '자동차 차종-연식-번호판 인식용 영상']  # root folder
dir_list = []

for f in f_list :
    dir_list.append(root_dir + f + '/')  # root folder 경로를 dir_list에 추가

ori_dir_list = dir_list.copy()  # 조회 할 경로 리스트
end_dir_list = []  # 하위 폴더 조회가 모두 끝난 경로만 모은 리스트
i = 0  # 반복문 조회 횟수를 담은 변수 : 경로 조회 진행 정도 파악용 

while True :
    append_list = []  # dir_list에 추가할 경로를 담을 리스트
    print(i, end=">>")  # 반복문 조회 횟수 출력
    
    # 경로 리스트의 경로를 하나씩 조회해보자.
    for dir0 in ori_dir_list :
        f_list = os.listdir(dir0)  # 경로의 폴더 리스트 조회
        # 조회된 파일리스트에서 폴더명만 리스트에 담기
        f_list2 = [f for f in f_list if (f[-4:] not in ignore_f + end_f) and (".DS_Store" not in f)]
        
        if 0 < len(f_list2) :  # 조회된 폴더가 있으면,
            append_list += [dir0 + f + '/' for f in f_list2]  # append_list에 경로를 추가
            dir_list.remove(dir0)  # 조회 완료된 경로는 제거
        elif len(f_list2) == 0 :  # 조회된 폴더가 없으면,
            end_dir_list.append(dir0)  # 조회가 완료된 경로로 end_dir_list에 추가
            dir_list.remove(dir0)  # 조회 완료된 경로는 제거
    dir_list += append_list  # dir_list에 새로 조회된 경로 추가

    if ori_dir_list == dir_list :  # 경로 조회 전과 경로 조회 후, 경로 리스트가 같으면,(새로 조회된 경로가 없으면)
        break  # 반복문 실행 중지
    else :  # 새로 조회된 경로가 있으면,
        ori_dir_list = dir_list.copy()  # ori_dir_list를 dir_list로 업데이트
        i += 1  # 반복문 조회 횟수 + 1
        
print()   
# 경로 조회 완료 후, 경로 리스트 출력
for f in end_dir_list :
    print(f)
````

# 폴더 딕셔너리 만들기

````python
f_dic = {'160. 차량파손 이미지 데이터':{}, '자동차 차종-연식-번호판 인식용 영상':{}}  # 상하위 폴더를 작성할 딕셔너리

for f in end_dir_list :
    folders = f.split('/')  # 경로를 '/'를 기준으로 쪼개서 리스트 생성
    
    # rootFolder의 인덱스 구하기
    try :
        i = folders.index('160. 차량파손 이미지 데이터')  
    except :
        i = folders.index('자동차 차종-연식-번호판 인식용 영상')
    folders.remove('')  # 폴더 리스트의 맨 마지막 공백 제거
    
    for j in range(i + 1, len(folders)) :
        now_dic = f_dic.copy()
        for k in range(i, j) :
            now_dic = now_dic[folders[k]]  # 경로의 인덱스k 폴더를 키 값으로 딕셔너리 값 조회
            if folders[j] not in now_dic :  # 현재 값에 경로의 인덱스 j 폴더가 없으면,
                now_dic[folders[j]] = {}  # 딕셔너리에 키 - 값 추가 : 인덱스k 폴더 - {} 추가


# 폴더 딕셔너리 출력
print(f_dic)
````

````python
from zipfile import ZipFile as zf
import numpy as np
from PIL import Image
from io import BytesIO
import os

zip_dic = {'/content/drive/MyDrive/Colab Notebooks/AIDA/Data/160. 차량파손 이미지 데이터/01.데이터/1.Training/1.원천데이터/': ['TS_damage_part.zip', 'TS_damage.zip'], '/content/drive/MyDrive/Colab Notebooks/AIDA/Data/160. 차량파손 이미지 데이터/01.데이터/2.Validation/1.원천데이터/': ['VS_damage_part.zip', 'VS_damage.zip']}

i = 0  # 압축 풀기 진행 상황 확인용 변수
for path in zip_dic.keys() :
  for z in zip_dic[path] :
    img_name = []  # 파일 이름을 담을 리스트
    img_arr_li = []  # 이미지 numpy array를 담을 리스트
    with zf(path + z, 'r') as rz :
      i += 1  # 경로 조회 진행도 + 1
      j = 0  # 파일 조회 진행도 초기화
      print(f'\n{i}번째 경로: { path + z }')  # 현재 조회중인 경로 출력
      
      f_list = rz.namelist()  # zip파일 내부 파일명 리스트
      
      for f in f_list :
        j += 1
        if f.endswith('.jpg') :
          print(f'\r진행도 : [{j} / {len(f_list)}] {j / len(f_list) * 100 : .2f}%', end="") # 현재 조회중인 zip파일 변환 진행도 출력
          img_f = rz.read(f)  # 이미지 파일 읽기 : bytes타입
          try :
            img = Image.open(BytesIO(img_f))  # bytes타입을 읽어와 이미지로 변수에 저장
            img_arr = np.array(img)  # 이미지를 numpy array로 변환
            
            # img_arr_li에 이미지가 존재하지 않으면,
            if sum([np.array_equal(arr,img_arr) for arr in img_arr_li]) == 0 :
              img_name.append(f)  # img_name에 이미지 파일 이름 추가
              img_arr_li.append(img_arr)  # img_arr_li에 이미지 numpy 배열 추가
            # 이미 리스트에 추가한 이미지라면,
            else :
              print('이미 존재하는 데이터')

          # bytes타입 변환 실패시,
          except UnicodeDecodeError:
            print("디코딩 에러")
          

      # 저장할 파일 이름 설정
      f_name = z[:-4]

      name = np.array(img_name)  # 이미지 파일 이름을 numpy 배열로 변환
      np.savez(f_name+'.npz', img_arr=img_arr_li, f_name=name)  # 이미지 numpy 배열을 압축한 npz파일 저장하기   

````
